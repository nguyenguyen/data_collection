```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/leanderme/sytora/master/data/all-files-for-ml/all_pivoted.csv')
df.tail()
```

```python
disease = pd.DataFrame(df['Disease'].values, columns=['Term'])
disease['Type'] = ['Disease']*len(disease)

symptom = pd.DataFrame(df.drop(columns=['Disease'], axis=1).columns, columns=['Term'])
symptom['Type'] = ['Symptom']*len(symptom)

print(len(symptom))
print(len(disease))

all_symp_dis = pd.concat([symptom, disease], axis=0, ignore_index=True)
all_symp_dis
```

```python
def build_adj_matrix(symp_map_mat):  
  num_disease = symp_map_mat.shape[0]
  num_symptom = symp_map_mat.shape[1]

  dis_empty_part = np.zeros((num_disease, num_disease))
  symp_empty_part = np.zeros((num_symptom, num_symptom))

  upper_part = np.concatenate((symp_empty_part, np.transpose(symp_map_mat)), axis=1)
  lower_part = np.concatenate((symp_map_mat, dis_empty_part), axis=1)

  return np.concatenate((upper_part, lower_part), axis=0)
```

```python
adj_mat = build_adj_matrix(df.drop(columns=['Disease'], axis=1).values)
```

```python
adj_mat = pd.DataFrame(adj_mat, index=all_symp_dis['Term'].values, columns=all_symp_dis['Term'].values)
```
